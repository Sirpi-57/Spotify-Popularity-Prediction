## Data Importing:

```python
import pandas as pd

df = pd.read_parquet('0000 (1).parquet')

print(df.head())
```

```python
df.columns
```

## Check For Null Values:

```python
df.isnull().sum()
```

```python
df.shape
```

```python
# Filter rows where the 'artists' column is null
null_artists = df[df['artists'].isnull()]

print(null_artists)
```

### This is the row that has all the 3 null columns. And hence dropped.

```python
# Remove rows where the 'artists' column is null (in-place)
df.dropna(subset=['artists'], inplace=True)

# Display the updated DataFrame
print(df)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.describe()
```

## Check For Duplicated rows:

```python
import pandas as pd

# Finding duplicate rows based on the 'track_id' column
duplicates = df[df.duplicated(subset=['track_id'], keep=False)]
print(duplicates)
```

```python
# Sorting the duplicate rows by 'track_id' to view them next to each other
sorted_duplicates = duplicates.sort_values(by='track_id')
print(sorted_duplicates)
```

```python
# Remove duplicate rows based on 'track_id', keeping only the first occurrence
df_cleaned = df.drop_duplicates(subset=['track_id'], keep='first')

# Display the cleaned DataFrame
print(df_cleaned)
```

```python
duplicates = df_cleaned[df_cleaned.duplicated(subset=['track_id'], keep=False)]
print(duplicates)
```
